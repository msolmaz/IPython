```python
inputFileData = TFile.Open("/Users/melihsolmaz/Desktop/ipython/Data22Jan_combined_2saMu.root")
nMinus1_looseCollCuts = inputFileData.Get("nMinus1_leptonD0Sig_vs_LxySig_looseCollCuts")
nMinus1_looseControlCollCuts = inputFileData.Get("nMinus1_leptonD0Sig_vs_LxySig_looseControlCollCuts")

#Define the function here....

def LxySig_distribution(y):
    final_region_dist = ROOT.TH1F("final_region_"+str(y), "", 24, 0, 12)
    control_region_dist = ROOT.TH1F("control_region_"+str(y), "", 24, 0, 12)

#Some Histogram Settings
    control_region_dist.GetXaxis().SetTitle("L_{xy}/#sigma_{L_{xy}}")
    control_region_dist.GetYaxis().SetTitle("Number of events with x < L_{xy}/#sigma_{L_{xy}} < 6 ")
    control_region_dist.SetTitle("#scale[0.8]{CMS Preliminary #sqrt{s} = 8 TeV, L = 20.5 fb^{-1}}")

    final_region_dist.SetStats(0)
    control_region_dist.SetStats(0)
    final_region_dist.SetMarkerStyle(8)
    control_region_dist.SetMarkerStyle(21)
    final_region_dist.SetMarkerColor(2)
    control_region_dist.SetMarkerColor(1)
    final_region_dist.SetLineColor(2)
    control_region_dist.SetLineColor(1)
    final_region_dist.Sumw2
    control_region_dist.Sumw2

    #control_region_dist.SetMaximum(400.0)


#nYbins= nMinus1_looseControlCollCuts.GetNbinsY()
#print "nYbins= ", nYbins

#nXbins= nMinus1_looseControlCollCuts.GetNbinsX()
#print "nXbins= ", nXbins

#Loop to write down new histos...

    for i in range(1,13):
        if y == 0 :
            final_region_dist.SetBinContent(i, nMinus1_looseCollCuts.Integral(2*y+1, 8, i, 12))
            control_region_dist.SetBinContent(i, nMinus1_looseControlCollCuts.Integral(2*y+1, 8, i, 12))
        else :
            final_region_dist.SetBinContent(i, nMinus1_looseCollCuts.Integral(2*y, 8, i, 12))
            control_region_dist.SetBinContent(i, nMinus1_looseControlCollCuts.Integral(2*y, 8, i, 12))
            
# Legend
    leg = TLegend(0.6,0.6,0.85,0.85)
    leg.SetFillColor(0)
    leg.SetLineColor(0)
    leg.AddEntry(final_region_dist, "Signal (|#Delta#Phi| < #pi/2)", "pl")
    leg.AddEntry(control_region_dist, "Control (|#Delta#Phi| > #pi/2)", "pl")

#Line
    line = TLine(6, 0, 6, 400)
    line.SetLineColor(4)
    line.SetLineStyle(2)

#Text
    text = TText(7, 2, "Excluded Region")

#print "total number of entries = ", control_region_dist.GetEntries()    
    canvas=TCanvas("canvas_"+str(y),"", 800, 800)
    control_region_dist.Draw("e")
    final_region_dist.Draw("esame")
    leg.Draw("same")
    line.Draw()
    text.Draw()
    canvas.Draw()
    canvas.SetLogy(True)
    return canvas, leg, final_region_dist, control_region_dist, text, line  
  
    

```

```python
def d0Sig_distribution(y):
    final_region_dist = ROOT.TH1F("final_region_d0_"+str(y), "", 24, 0, 12)
    control_region_dist = ROOT.TH1F("control_region_d0_"+str(y), "", 24, 0, 12)

#Some Histogram Settings
    control_region_dist.GetXaxis().SetTitle("|d_{0}|/#sigma_{d_{0}}")
    control_region_dist.GetYaxis().SetTitle("Number of events with x < d_{0}/#sigma_{d_{0}} < 4 ")
    control_region_dist.SetTitle("#scale[0.8]{CMS Preliminary #sqrt{s} = 8 TeV, L = 20.5 fb^{-1}}")

    final_region_dist.SetStats(0)
    control_region_dist.SetStats(0)
    final_region_dist.SetMarkerStyle(8)
    control_region_dist.SetMarkerStyle(21)
    final_region_dist.SetMarkerColor(2)
    control_region_dist.SetMarkerColor(1)
    final_region_dist.SetLineColor(2)
    control_region_dist.SetLineColor(1)
    final_region_dist.Sumw2
    control_region_dist.Sumw2

    #control_region_dist.SetMaximum(400.0)


#nYbins= nMinus1_looseControlCollCuts.GetNbinsY()
#print "nYbins= ", nYbins

#nXbins= nMinus1_looseControlCollCuts.GetNbinsX()
#print "nXbins= ", nXbins

#Loop to write down new histos...

    for i in range(1, 8):
        if y == 0 :
            final_region_dist.SetBinContent(i, nMinus1_looseCollCuts.Integral(i, 8, 2*y+1, 12))
            control_region_dist.SetBinContent(i, nMinus1_looseControlCollCuts.Integral(i, 8, 2*y+1, 12))
        else :
            final_region_dist.SetBinContent(i, nMinus1_looseCollCuts.Integral(i, 8, 2*y, 12))
            control_region_dist.SetBinContent(i, nMinus1_looseControlCollCuts.Integral(i, 8, 2*y, 12))
            
# Legend
    leg = TLegend(0.6,0.6,0.85,0.85)
    leg.SetFillColor(0)
    leg.SetLineColor(0)
    leg.AddEntry(final_region_dist, "Signal (|#Delta#Phi| < #pi/2)", "pl")
    leg.AddEntry(control_region_dist, "Control (|#Delta#Phi| > #pi/2)", "pl")

#Line
    line = TLine(4, 0, 4, 400)
    line.SetLineColor(4)
    line.SetLineStyle(2)

#Text
    text = TText(5, 2, "Excluded Region")

#print "total number of entries = ", control_region_dist.GetEntries()    
    canvas=TCanvas("canvas_d0_"+str(y),"", 800, 800)
    control_region_dist.Draw("e")
    final_region_dist.Draw("esame")
    leg.Draw("same")
    line.Draw()
    text.Draw()
    canvas.Draw()
    canvas.SetLogy(True)
    return canvas, leg, final_region_dist, control_region_dist, text, line  
```
