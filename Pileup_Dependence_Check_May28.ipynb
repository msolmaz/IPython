```python
import ROOT
from math import sqrt
from ROOT import TFile, TDirectory, TH1F, TCanvas, TLegend, TLine, TText, TPad, TF1, TEfficiency, \
                 TGraph, TGraphAsymmErrors

import rootnotes
import rootprint
ROOT.gStyle.SetErrorX(0)
```

```python
file1 = TFile.Open("HTo2LongLivedTo4F_MH1000_MFF350_CTau35To3500_analysis_20140504_weighted_2saMu.root", "READ")

histo_cosmic_before = file1.Get("before_cosmic_pile_up_finalCollCuts")
histo_cosmic_after = file1.Get("after_cosmic_pile_up_finalCollCuts")

histo_track_before = file1.Get("before_track_pile_up_finalCollCuts")
histo_track_after = file1.Get("after_track_pile_up_finalCollCuts")

histo_cosmic_before.Rebin(5)
histo_cosmic_after.Rebin(5)

histo_track_before.Rebin(5)
histo_track_after.Rebin(5)

eff1 = TH1F("eff1", "", 10, 0, 50)
eff2 = TH1F("eff2", "", 10, 0, 50)

eff1.GetXaxis().SetTitle("Number of Reconstructed Primary Vertices")
eff1.GetYaxis().SetTitle("Cosmic Rejection Efficiency")

eff2.GetXaxis().SetTitle("Number of Reconstructed Primary Vertices")
eff2.GetYaxis().SetTitle("Track Rejection Efficiency")

eff1.SetStats(0)
eff2.SetStats(0)
eff1.SetMarkerStyle(8)
eff2.SetMarkerStyle(21)
eff1.SetLineColor(2)
eff2.SetLineColor(1)
```

```python
for i in range(1, 10):
    if histo_cosmic_before.GetBinContent(i) == 0:
        continue
    eff1.SetBinContent(i, histo_cosmic_after.GetBinContent(i)/histo_cosmic_before.GetBinContent(i))
    
for i in range(1, 10):
    if histo_track_before.GetBinContent(i) == 0:
        continue
    eff2.SetBinContent(i, histo_track_after.GetBinContent(i)/histo_track_before.GetBinContent(i))
    
```

```python
c1 = TCanvas()
eff1.Draw()
c1.Draw()
c1
```

```python
c2 = TCanvas()
eff2.Draw()
c2.Draw()
c2
```
