```python
import ROOT
import rootnotes
import rootprint
import math
from ROOT import RooRealVar, RooFormulaVar, RooVoigtian, RooChebychev, RooArgList, \
                 RooArgSet, RooAddPdf, RooDataSet, RooCategory, RooSimultaneous, \
                 RooBreitWigner, RooCBShape, RooFFTConvPdf, TFile, TDirectory, \
                 TH1F, TCanvas, TLegend, TLine, TText, TPad, TF1, RooDataHist, RooHist, \
                 RooPlot

ROOT.gStyle.SetErrorX(0)
```

```python
inputMC = TFile.Open("MC_Z_width.root")
histoMC = inputMC.Get("Mass_looseControlCollCuts")

#To account for weight of DrellYan events...
histoMC.Scale(3503.7)
```

```python
#Parameter of the histogram
mass = RooRealVar("mass", "mass", 0, 500)

rooMC = RooDataHist("rooMC", "Di Muon Mass Distribution DY Jets", RooArgList(mass), histoMC)
```

```python
#Parameters of fit function
mean = RooRealVar("mean", "mean", 60, 0, 130)
width = RooRealVar("width", "width", 6, 1, 40)
sigma = RooRealVar("sigma", "sigma", 5, 0.2, 40)
alpha = RooRealVar("alpha", "#alpha", -0.1, -10.0, -0.1)
npow = RooRealVar("npow", "n_{CB}", 2.3, 0.1, 10.)

#PDF's
pdfMC = RooVoigtian("pdfMC", "pdfMC", mass, mean, width, sigma)
pdfMC2 = RooCBShape("pdfMC2", "pdfMC2", mass, mean, width, alpha, npow)
```

```python
frameMC = mass.frame(0, 250)

rooMC.plotOn(frameMC)

pdfMC2.fitTo(rooMC)

pdfMC2.plotOn(frameMC)
```

Z-mass fit with CB function for DY Jets

```python
frameMC.GetXaxis().SetTitle("Mass[GeV/c^2]")

c1 = TCanvas("c1", "Canvas", 500, 500)
frameMC.Draw()
c1
```

Z-mass fit with Voigtian function for DY Jets

```python
frameMC2 = mass.frame(0, 250)

rooMC.plotOn(frameMC2)

pdfMC.fitTo(rooMC)

pdfMC.plotOn(frameMC2)

c2 = TCanvas("c2", "Canvas", 500, 500)
frameMC2.Draw()
c2
```
